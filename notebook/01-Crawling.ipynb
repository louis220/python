- 데이터를 크롤링하여 정규식으로 몇 가지 사실을 알아내 봅시다


```python
# 수집 대신 url
url = "http://www.gutenberg.org/cache/epub/12429/pg12429.txt"
#urllib.loquest 객체 필요
import urllib.request
# 페이지 데이터 가져오기 : urlopen
page = urllib.request.urlopen(url)
# print(page, type(page))
# 페이지의 텍스트 불러오기 : read
page_text = page.read().decode("utf-8")
print("page text:", page_text)
# 디코딩 :decode()
# page_text = page_text.decode("utf-8")
print(page_text[:100])
```

```python
# #0124(#으로 시작하고 숫자가 붙은 형태) -> 뭔가의 코드일 것이다
import re
# 아마도 이것은 뭔가의 코드일 것이다
iter = re.finditer(r"#[0-9]+", page_text)
# fineiter -> SRE_Watch 객체의 리스트로 저장
# 순회할 수 있다
for code in iter:
    print("code:", code.group())
```

```python
# 수집한 데이터가 lasr leaf니까
# leaf 라는 단어가 얼마나 나오는지 
print("leaf:", re.findall(r"leaf", page_text))
# leaves는 얼마나 나오나? (복수형)
print("leaves", re.findall(r"leaves", page_text))
# Leaf or Leaves는 얼마나 나오나?
print("Leaf or leaves:", re.findall(r"Leaf|Leaves", page_text))

# 한번에 뽑아낼 수 있는 패턴은 없을까?
print(re.findall("leaf|leaves", page_text, re.IGNORECASE))

# 첫 글자가 대문자인 경우 - 고유명사이거나 문장의 첫 단어일 가능성이 있다
p = re.compile(r"\b[A-Z]{1}[a-z]+")
iter = re.finditer(p, page_text)

# for word in iter:
#     print("첫 글자가 대문자인 단어:", word.group())
    
# 첫 글자가 대문자인 단어의 list 생성
words = [item.group().strip() for item in iter]
# print("words",words)
# 정렬
words.sort()
# print(words)
# 중복 제거
words= list(set(words))
print("첫 글자가 대문자인 단어의 갯수:", len(words))
```
