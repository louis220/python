```python
# 사이킷 런을 이용한 붓꽃 분류 문제
# 사이킷 런의 k-최근접 이웃 알고리즘을 사용
# 데이터 셋 로드
from sklearn.datasets import load_iris
iris = load_iris()
print("type of iris:", type(iris))
print("keys of iris dataset:", iris.keys())

print("target names:", iris['target_names']) # 분류명
# 결과가 
# 0 :setosa, 1:versicolor, 3: virginica
```

```python
# 데이터셋의 설명
print(iris['DESCR'])
```

```python
# feature의 목록
iris.feature_names
```

```python
# data 크기 확인
print("Size of Data:", iris.data.shape) # 150행 4열
iris.data[:10]
```

```python
# target의 크기 확인
print(iris.target.shape) # 150행 1차원 벡터
print(iris.target[:10])
# target의 이름을 확인
print(iris.target_names)
```

```python
# 학습 데이터와 테스트 데이터를 분할
# 일반적으로 8: 2 혹은 7 :3 정도로 분할
# 테스트 데이터는 학습에 절대로 사용해서는 안된다
# 사이킷 런의 데이터 분할 함수를 이용해서 데이터셋 분리
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=0)
# 데이터를 확인 
print("학습 데이터 크기:", x_train.shape)
print("테스트 데이터 크기:", x_test.shape)
print("학습 타겟 데이터 크기:", y_train.shape)
print("테스트 타겟 데이터 크기:", y_test.shape)

# 전체 (학습)데이터를 출력
for _x, _y in zip(x_train, y_train):
    print("학습 데이터{} -> 타겟 데이터 {} -> 품종 {}".format(_x, _y,iris.target_names[_y]))
```

```python
# 데이터 살표보기
# 정규화 필요, 결측치 등의 선처리가 필요
import pandas as pd
df_iris = pd.DataFrame(x_train, columns=iris.feature_names)
df_iris.head()
```

```python
%matplotlib inline
# 데이터 분포를 시각화 :scatter_matrix
import matplotlib.pyplot as plt
pd.plotting.scatter_matrix(df_iris, c= y_train, figsize=(15, 15),marker="o")
plt.show()
```

```python
# 사이킷 런의 k-최근접 이웃 알고리즘을 이요해 학습
from sklearn.neighbors import KNeighborsClassifier

# 모델 만들기
knn = KNeighborsClassifier(n_neighbors=1)
# 학습 수행 :fit
knn.fit(x_train, y_train)
```

```python
# 만들어진 모델로 새로운 데이터를 예측 ->predict
import numpy as np
x_new = np.array([[5, 2.9, 1, 0.2]]) # ->어느 품종?
#예측
prediction = knn.predict(x_new)
print("예측", prediction)
print("새로운 데이터 {}, 예측 품종{}".format(x_new, iris.target_names[prediction]))
```

```python
# 모델 평가
# 만들어진 모델에 테스트 데이터와 테스트 타겟으로 검증
y_pred = knn.predict(x_test)
print("테스트 데이터에 대한 예측치")
for pred in y_pred:
    print("테스트 데이터 예측치 {} -> 예측 품종 {}".format(pred, iris.target_names[pred]))
```

```python
# 모델에 대한 정확도 계산, 모델의 성능 평가
print("테스트 세트의 정확도:{}".format(knn.score(x_test, y_test)))
print("테스트 세트의 정확도:{}".format(np.mean(y_pred == y_test)))
```
